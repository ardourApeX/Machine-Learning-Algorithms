## Closed form Solution in Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
# Generate Dataset

X, Y = make_regression(n_samples = 400, n_features = 1, n_informative = 1, noise = 1.8, random_state = 11)
```

```python
# Y is 1d

Y = Y.reshape((-1, 1)) # Now its a vector

X.shape, Y.shape
```

## Adding Bias Column

```python
ones = np.ones((X.shape[0], 1))

X_= np.hstack((ones, X))
```

```python
X_.shape
```

```python
def predict(X, theta):
    return np.dot(X, theta)
```

$ \theta = (X^T X)^{-1} X^{T}Y$

```python
def get_theta_closed_form(X, Y):
    
    Y = np.mat(Y)
    product1 = np.dot(X.T, X)
    product1 = np.linalg.pinv(product1) # Taking Pseudo Inverse
    product2 = np.dot(X.T, Y)
    
    return product1 * product2

    
```

```python
theta = get_theta_closed_form(X, Y)
```

```python
theta
```

```python
Y_predict = predict(X, theta)
Y_predict.shape, X.shape
```

```python
plt.style.use("seaborn")
plt.scatter(X,Y, alpha = 0.7, label = "Actual")
plt.plot(X, Y_predict, color = 'orange', label = "Predicted") 
plt.legend()
plt.title("Normalized Data");
```
