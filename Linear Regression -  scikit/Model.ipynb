```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression #this will provide sample dataset
```

## Dataset

```python
X, Y = make_regression(n_samples = 1000) # Explicitly defining no of samples otherwise it would be 100

X.shape, Y.shape
```

```python

X, Y = make_regression(n_samples = 1000, n_features = 2, n_informative = 1, noise = 10, random_state = 1)
X.shape, Y.shape
```

## Unuseful Features

```python
# Now from the above sample there is a feature which isn't usefull because n_informative = 1
# Lets find it out
plt.style.use("seaborn")
plt.subplot(1, 2, 1) 
plt.scatter(X[:, 0], Y) # First Feature

plt.subplot(1, 2, 2) 
plt.scatter(X[:, 1], Y, color = 'orange') # Second Feature
plt.plot();
```

## Visiualization : All Features at once

```python
from mpl_toolkits import mplot3d
%matplotlib notebook


fig = plt.figure(figsize = (5, 7)) # Figure size
ax = plt.axes(projection = '3d') # Creating an object of 3D axes

# Creating a 3D plot

ax.scatter3D(X[:, 0], X[:, 1], Y, color = 'blue')
plt.title("3D scatter plot")
plt.show()
```

## Implementation

```python
from sklearn.linear_model import LinearRegression # Package for Linear Regression
```

```python
model = LinearRegression() #Object
```

```python
model.fit(X, Y)
```

```python
# Accessing Theta : 

model.coef_ # It holds theta from 1 to n i.e. theta1, theta2, theta3 ... thetan
```

```python
# Theta0 

model.intercept_
```

## Prediction

```python
model.predict([X[0], X[1]])
```

```python
print(Y[0], Y[1])
```

## Model Accuracy

```python
model.score(X, Y)
```
