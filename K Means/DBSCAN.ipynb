### DBSCAN : Density Based Spatial Aplications & Noise

````python
# Generate Dataset : This time it will be a little tricky

import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_circles
````

````python
X, Y = make_moons(n_samples = 200, noise = 0.10)
````

````python
plt.scatter(X[:, 0], X[:, 1], c = Y)
plt.show()
````

# Implementing K-Means

````python
from sklearn.cluster import KMeans
````

````python
kmeans = KMeans(n_clusters = 2)
````

````python
kmeans.fit(X)
````

````python
centers = kmeans.cluster_centers_
````

````python
Predict = kmeans.labels_
````

````python
plt.scatter(X[:, 0], X[:, 1], c = Predict)
plt.scatter(centers[:, 0], centers[:, 1], c = "red", marker = "*")
plt.show()
````

````python
from sklearn.cluster import DBSCAN
dbs = DBSCAN(eps = 0.1, min_samples = 5)
````

````python
dbs.fit(X)
````

````python
pred = dbs.fit_predict(X)
````

````python
plt.scatter(X[:, 0], X[:, 1], c = pred)
#plt.scatter(centers[:, 0], centers[:, 1], c = "red", marker = "*")
plt.show()
````

````python
# Here you can see, DBSCAN automatically made more than 2 cluster center points

dbs = DBSCAN(eps = 0.3, min_samples = 5)
pred = dbs.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c = pred)
#plt.scatter(centers[:, 0], centers[:, 1], c = "red", marker = "*")
plt.show()
````
